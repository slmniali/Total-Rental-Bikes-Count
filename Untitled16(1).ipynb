```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
```

```python
df = pd.read_csv('/content/hour.csv')
```

```python
df.head()
```

```python
dummy = ['season', 'weathersit', 'mnth', 'hr', 'weekday']
for i in dummy:
    dummies = pd.get_dummies(df[i], prefix=i, drop_first=False)
    df = pd.concat([df, dummies], axis=1)

fields_to_drop = ['instant', 'dteday', 'season', 'weathersit',
                  'weekday', 'mnth', 'hr', 'workingday', 'casual', 'registered']
df = df.drop(fields_to_drop, axis=1)
df.head()
```

```python
scaler = StandardScaler()
```

```python
columns = ['temp', 'atemp', 'hum']
df[columns] = scaler.fit_transform(df[columns])
df.head()
```

```python
test_df = df[-73*24:]
train_df = df[:-73*24]

targets = ['cnt']

X_train = train_df.drop(targets, axis=1)
Y_train = train_df[targets]

X_test = test_df.drop(targets, axis=1)
Y_test = test_df[targets]
```

```python
Reg = LinearRegression()
```

```python
Reg_Model = Reg.fit(X_train, Y_train)
Predict = Reg_Model.predict(X_test)
r_squared = r2_score(Y_test, Predict)
print('The r_squared of this model is:', r_squared)
```

```python
total_predicted_count = Predict.sum()
total_actual_count = Y_test.sum()
print('The total predicted count of rental bikes is:', total_predicted_count)
print('The total actual count of rental bikes is:', total_actual_count)
```
